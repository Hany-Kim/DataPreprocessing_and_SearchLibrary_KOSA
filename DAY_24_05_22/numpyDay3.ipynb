### 리스트를 이용해 데이터프레임 만들기

2개 이상의 리스트를 이용해 데이터프레임 생성하는 법

```python
L1 = [1,2,3,4,5]
L2 = [6,7,8,9,10]
```

```python
df = pd.DataFrame(data={'col1':L1, 'col2':L2})
df
```

```python
import numpy as np
```

c_[]로 리스트를 열 단위로 연결 후\
columns로 열 이름 지정, 지정하지 않으면 0열부터 순서대로 자동생성

```python
df = pd.DataFrame(np.c_[L1, L2],
                  columns=['col1','col2'])
df
```

```python
df = pd.DataFrame(np.c_[L1, L2])
df
```

```python
df = pd.DataFrame(np.r_[L1, L2],
                  index=['row1','row2','row3','row4','row5',
                         'row6','row7','row8','row9','row10'])
df
```

L1, L2가 리스트의 리스트(2차원 리스트)인 경우

```python
L1 = [[1,2,3,4,5]]
L2 = [[6,7,8,9,10]]
```

c_[]를 사용하면 1개의 행으로 표현된다.

```python
df = pd.DataFrame(np.c_[L1,L2])
df
```

r_[]를 사용하면 행단위로 리스트를 쌓는다.

```python
df = pd.DataFrame(np.r_[L1,L2])
df
```

```python
L1 = [[1,2,3],[6,7,8]]
L2 = [[8,9,10],[11,12,13]]
```

```python
df = pd.DataFrame(np.c_[L1,L2])
df
```

```python
df = pd.DataFrame(np.r_[L1,L2])
df
```

### iris 데이터셋
붓꽃의 3종류
> setosa(세토사) | versicolor(버시컬러) | virginica(버지니카)
* 각 종류 별로 50개씩 Data를 모아둠
* 각 Data의 정보 종류로는
    * Sepal length & width => 꽃받침 길이 & 너비
    * Petal length & width => 꽃잎 길이 & 너비
    * Species => 종

#### Scikit-learn패키지의 iris 데이터셋

```python
from sklearn import datasets
```

* iris 데이터는 딕셔너리 형태
    * key값 = {data, target, target_names, frame,  DESCR, feature_names, filename, data_module, ...}
    * data => 독립변수 : 어떤 것에 영향을 주는 변수
    * target => 종속변수 : 어떤 것으로부터 영향을 받는 변수

```python
iris_dic = datasets.load_iris()
iris_dic
```

data와 target을 이용해 x(독립변수), y(종속변수)를 DataFrame으로 만듬

```python
x = pd.DataFrame(iris_dic.data, columns=iris_dic.feature_names)
y = pd.DataFrame(iris_dic.target_names[iris_dic.target], columns=['species'])
```

```python
print(x)
print(y)
```

위 코드를 딕셔너리를 참조하는 형태로 []를 사용해 같은 결과값을 출력

```python
x = pd.DataFrame(iris_dic['data'], columns=iris_dic['feature_names'])
y = pd.DataFrame(iris_dic['target_names'][iris_dic['target']], columns=['species'])
```

concat()으로 x와 y를 붙여서 하나의 DataFrame으로 생성\
axis=1(열,오른쪽)에 붙임

```python
iris = pd.concat([x,y], axis=1)
iris.head()
```

concat()으로 x와 y를 붙여서 하나의 DataFrame으로 생성\
axis=0(행,아래)에 붙임\
원소가 비어있는 곳은 NaN

```python
iris = pd.concat([x,y], axis=0)
iris
```

species에 이름이 아닌 숫자 값을 가지게 함\
setosa => 0\
versicolor => 1\
viginica => 2

```python
y = pd.DataFrame(iris_dic.target, columns=['species'])
iris = pd.concat([x,y], axis=1)
```

```python
#iris.iloc[:, :-1] # 독립변수만 뽑기, 마지막 열빼고 뽑기
A = iris.iloc[:, :-1].to_numpy()
np.var(A), A.var()
```

```python
np.var(A, axis=0), A.var(axis=0) # 각 열별 분산
```

numpy에도 분산이 있다.\
numpy와 pandas간 분산 구하는 방법이 다르다.\
numpy로 구한 분산 값 : 0.68112222\
pandas로 구한 분산 값 : 0.685694\
결과값이 다르게 나온다.\
\
ddof 매개변수 (델타 ~~ )를 1로 두면 같은 결과값이 나온다.

```python
np.var(A, axis=0), A.var(axis=0, ddof=1)
```
