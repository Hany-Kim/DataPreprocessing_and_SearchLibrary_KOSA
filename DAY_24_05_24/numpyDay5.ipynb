2.3.2. subplots() 함수로 서브플롯 집합 추가

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10, 0.01)
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0,0].plot(x, np.sin(x)); axes[0,1].plot(x, np.cos(x))
axes[1,0].plot(x, np.tanh(x)); axes[1,1].plot(x, np.sin(x)*np.cos(x))
plt.show()
```

```python
x = np.arange(0, 7, 0.01)
```

```python
def sin_cos(x):
    return np.sin(x)*np.cos(x)
```

```python
func_list = [np.sin, np.cos, np.tanh, sin_cos]
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
for i, ax in enumerate(axes.flat):
    ax.plot(x, func_list[i](x))
plt.show()
```

```python
print(fig, axes)
```

```python
fig, axes = plt.subplots(ncols=4)
for i, ax in enumerate(axes.flat):
    ax.plot(x, func_list[i](x))
plt.show()
```

```python
fig, axes = plt.subplots(nrows=4)
for i, ax in enumerate(axes.flat):
    ax.plot(x, func_list[i](x))
plt.show()
```

2.4.2 plot()

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10, 0.6)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8,5))
axes[0,0].plot(y)
axes[0,1].plot(x,y,'ro:')
axes[1,0].plot(x,np.cos(x),'cs-.')
axes[1,1].plot([3,6], [3,5], 'b^:')
axes[1,1].plot([4,5], [5,4], 'kx--')
plt.show()
```

2.4.3. scatter()

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
plt.scatter(x=iris.petal_length, y=iris.petal_width,
            s=iris.sepal_length*10, c=iris.sepal_width,
            alpha=0.5)
plt.show()
```

```python
np.random.seed(7902)
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2
```

2.4.4. bar(), barh(), axvline(), axhline()

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0,0].bar([1,2,3],[3,4,5])
axes[0,1].barh([1,3,5],[1,2,3])
axes[1,0].axvline(0.6)
axes[1,1].axhline(0.4)
plt.show()
```

2.4.5. hist()

```python
fig, axes = plt.subplots(2,2)
axes[0,0].hist(iris.sepal_length, bins=10, color='r')
axes[0,1].hist(iris.sepal_length, bins=15, cumulative=True)
axes[1,0].hist(iris.sepal_length, bins=5, orientation='horizontal')
axes[1,1].hist(iris.sepal_length, bins=10, histtype='step')
plt.show()
```

2.4.6. boxplot(), violineplot()

```python
fig, axes = plt.subplots(1,2)
axes[0].boxplot(iris.sepal_length)
axes[1].violinplot(iris.sepal_length)
plt.show()
```

2.4.7. fill(), fill_between(), fill_betweenx()

```python
x = np.linspace(0, 10, 50)
y = np.cos(x)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10,5))
axes[0].fill(x, y, c='blue')
axes[1].fill_between(x,y,color='red')
axes[2].fill_between(x,y,color='green')
plt.show()
```
