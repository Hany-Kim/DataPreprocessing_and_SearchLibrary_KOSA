\
IQR = Q3 (3/4 분위수, 75%) - Q1 를 구할 때 사용

```python
plt.boxplot(iris.sepal_length)
plt.show()
```

```python
iris.describe()
```

이상치 분석

```python
Q3 = 3.300000
Q1 = 2.800000
IQR = Q3 - Q1
upper = Q3 + 1.5*IQR
lower = Q1 - 1.5*IQR
```

```python
print(lower)
```

```python
iris.loc[iris.sepal_width < lower, :]
```

```python
iris.loc[(iris.sepal_width < lower) | (iris.sepal_width>upper), :]
```

## 2절. Matplotlib을 이용한 시각화

### 2.5. 그래프 커스터마이징

1) linestyle, linewidth

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.cos(x)
```

```python
fig, axes = plt.subplots(2,2) # 가로 2개, 세로 2개로 각각 나눔
axes[0,0].plot(x, y, linewidth=2, color='#FF00FF') # 색을 16진수로 표현
axes[0,1].plot(x, y, linestyle='dotted', linewidth=7, color='purple')
axes[1,0].plot(x, y, ls='--', lw=3, c='r') # 매개 변수 축약 가능
axes[1,1].plot(x, y, 'b, ') # '색점선' 형식으로 지정, 1픽셀 파란색
plt.show()
```

2) text(), annotate()

```python
fig, axes = plt.subplots(1,2)
axes[0].scatter(x, y, marker='.')
axes[0].text(2, 0, 'Example Graph', style='italic')
axes[1].scatter(x, y, marker='*')
axes[1].annotate('Sine', xy=(5, 0.5), xytext=(2, 0.75),
                 arrowprops=dict(arrowstyle='->', connectionstyle='angle3'))
plt.show()
```

4) 축과 눈금

```python
x = np.linspace(0, 10, 100)
y = np.cos(x)
```

```python
fix, axes = plt.subplots(2,2, figsize=(8,5))
plt.subplots_adjust(hspace=0.4, wspace=0.3)
axes[0,0].scatter(x, y, marker='.')
axes[0,0].set(title='An Example Axes', ylabel='Y-Axis', xlabel='X-Axis')

axes[0,1].scatter(x, y, marker='^', c='r')
axes[0,1].set_xlim(0,5)
axes[0,1].set_ylim(-2,2)
axes[0,1].set_xlabel('X 0~5')
axes[0,1].set_ylabel('Y -2~2')

axes[1,0].scatter(x, y, marker='.', c='c')
axes[1,0].set(xticks=range(1,8,2),
              xticklabels=[3,100,-12,'foo'],
              yticks=[-2,0,1,2],
              yticklabels=[-200,0,100,'Max'])

axes[1,1].scatter(x, y, marker='v', c='b')
axes[1,1].set_xticks(range(1,8,2))
axes[1,1].set_xticklabels([3,100,-12,'foo'])
axes[1,1].set_yticks([-2,0,1,2])
axes[1,1].set_yticklabels([-200,0,100,'Max'])
axes[1,1].spines['top'].set_visible(False)
axes[1,1].spines['bottom'].set_position(('outward', 10))
axes[1,1].grid(True)
plt.show()
```

5) 축 공유

```python
x = np.arange(0, 10)
y1 = 0.5 * x**2
y2 = -1 * y1
```

```python
fig, ax1 = plt.subplots()
ax1.plot(x, y1, 'g^:')
ax1.set_xlabel('X data')
ax1.set_ylabel('Y1 data', color='g')

ax2 = ax1.twinx()
ax2.plot(x, y2, 'bv--')
ax2.set_ylabel('Y2 data', color='b')

ax3 = ax1.twiny()
ax3.plot(-x, y1, 'ro-.')
ax3.set_xlabel('-X data', color='r')

plt.show()
```

```python
from cycler import cycler
my_cycler = (cycler('color', ['r','g','b','c','m','y','k']) + \
             cycler(linestyle=['-','--',':','-.','-','--',':']) + \
             cycler(lw=np.linspace(5, 20, 7)))
plt.rcParams['axes.prop_cycle'] = my_cycler
plt.plot([1,2], [1,1]); plt.plot([1,2],[2,2]); plt.plot([1,2],[3,3])
plt.plot([1,2], [4,4]); plt.plot([1,2],[5,5]); plt.plot([1,2],[6,6])
plt.plot([1,2], [7,7]); plt.plot([1,2],[8,8]); plt.show()
```
