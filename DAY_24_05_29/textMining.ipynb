```python
import requests
url = 'https://www.yonhapnewstv.co.kr/category/news/politics/feed/'
news = requests.get(url)
print(news)
```

```python
if news.status_code == 200:
    print('연합뉴스 기사를 로드했습니다.')
    from bs4 import BeautifulSoup
    news_soup = BeautifulSoup(news.content, "xml")
    links = news_soup.select("item > link")

    from konlpy.tag import Kkma
    kkma = Kkma()
    news_nouns_list = []
    for link in links:
        news_url = link.text
        news_response = requests.get(news_url)
        news_soup = BeautifulSoup(news_response.content, 'html.parser')
        news_content = news_soup.select_one('.cont-item01 > .detail')
        news_nouns = kkma.nouns(news_content.text)
        news_nouns_filter = filter(lambda word : len(word)>1 , news_nouns)
        news_nouns_list.append(list(news_nouns_filter))
else :
    print('기사를 로드하는데 실패했습니다.')
```

```python
from itertools import chain
text = ' '.join(list(chain(*news_nouns_list)))
```

```python
import numpy as np
from PIL import Image
img = Image.open('south_korea.png').convert('RGBA')
mask_ar = np.array(img)
stop_w = {"연합뉴스", "연합", "기사문의", "문의", "앵커", "기자", "뉴스", "제보", "닫기", "기사", "보도", "라인", "오늘", "시간"}
```

```python
import wordcloud as wc
wordc = wc.WordCloud(font_path="C:\\Windows\\Fonts\\malgun.ttf",
                     stopwords=stop_w, max_words=2000, mask=mask_ar,
                     background_color="white")
wordc.generate(text)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,15))
plt.imshow(wordc)
plt.axis('off')
plt.show()
```
